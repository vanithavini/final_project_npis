```python
# Load the data
import pandas as pd

df = pd.read_csv("../data/processed/news_cleaned.csv")
print(df.shape) 
```

```python
# Initialize Scorer
from src.popularity_scorer import PopularityScorer
scorer = PopularityScorer() 
```

```python
# Sanitize text column
df["title"] = df["title"].fillna("").astype(str)
df["description"] = df["description"].fillna("").astype(str) 
```

```python
# Drop rows that are completely Empty
df = df[(df["title"].str.strip() != "") | (df["description"].str.strip() != "")] 
```

```python
# Apply Popularity Scoring(row-wise)
scores = []
percentages = []
labels = []

for text in df["title"] + ". " + df["description"]:
    final_score, components = scorer.score_article(text)
    label = scorer.priority_label(final_score)

    scores.append(final_score)
    percentages.append(final_score)  # already 0–100
    labels.append(label)

df["popularity_score"] = scores
df["popularity_percentage"] = percentages
df["priority_label"] = labels 
```

```python
# Save popularity scored data
output_path = "../data/processed/news_popularity_scored.csv"
df.to_csv(output_path, index=False)

print(f"✅ Popularity scored file saved at: {output_path}") 
```
