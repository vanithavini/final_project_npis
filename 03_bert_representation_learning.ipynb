```python
# Import and Path setup
import os
import sys
import pandas as pd
from tqdm import tqdm

PROJECT_ROOT = os.path.abspath("..")
sys.path.append(PROJECT_ROOT)

print("Project root added:", PROJECT_ROOT) 
```

```python
# Load cleaned data
data_path = r"C:\Vanitha\Documents\GUVI\final_project\news_popularity_system\data\processed\news_cleaned.csv"
df = pd.read_csv(data_path)

print(df.shape) 
df.head() 
```

```python
# Prepare Text input
df["combined_text"] = df["title"].astype(str) + ". " + df["description"].astype(str)
texts = df["combined_text"].tolist()

print("Total documents:", len(texts))  
```

```python
# Initialize Embedding Generator
from src.embedding_generator import BertEmbeddingGenerator

embedder = BertEmbeddingGenerator(
    model_name="distilbert-base-uncased",
    max_length=256,
    batch_size=8  # ‚¨Ö safe for CPU & RAM
) 
```

```python
# Generate Embeddings in Chunks
embedder.generate_embeddings_in_chunks(
    texts=texts,
    save_dir="../data/processed/embeddings_chunks",
    chunk_size=5000,     # ‚¨Ö VERY IMPORTANT
    start_chunk=0
)  
```

```python
import numpy as np

chunk_dir = "../data/processed/embeddings_chunks"
files = sorted([f for f in os.listdir(chunk_dir) if f.endswith(".npy")])

embeddings = [np.load(os.path.join(chunk_dir, f)) for f in files]
final_embeddings = np.vstack(embeddings)

np.save("../data/processed/news_embeddings.npy", final_embeddings) 
```
