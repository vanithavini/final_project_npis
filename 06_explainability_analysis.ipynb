```python
# Import and path setup
import os
import sys

PROJECT_ROOT = os.path.abspath("..")
sys.path.append(PROJECT_ROOT)

print("Project root added:", PROJECT_ROOT) 
```

```python
# Load Dependencies & Data
import pandas as pd
from src.popularity_scorer import PopularityScorer
from src.explainability_engine import ExplainabilityEngine

df = pd.read_csv("../data/processed/news_popularity_scored.csv") 
```

```python
# Initialize Engines
scorer = PopularityScorer()
explainer = ExplainabilityEngine(scorer.weights) 
```

```python
# Pick a Sample Article (Top-ranked)
sample = df.sort_values("popularity_percentage", ascending=False).iloc[0]
text = sample["title"] + ". " + sample["description"] 
```

```python
# Recompute Signal Scores for Explanation
final_score, signal_scores = scorer.score_article(text)
contributions = explainer.explain(signal_scores)
top_reasons = explainer.top_reasons(contributions) 
```

```python
# Display Explanation
print("üì∞ Article Title:")
print(sample["title"])

print("\nüìä Popularity Score:", round(final_score, 2))

print("\nüîç Top Contributing Factors:")
for signal, value in top_reasons:
    print(f"- {signal.capitalize()}: {value}%")  
```

```python
print("\nüìä Full Contribution Breakdown:")
for signal, value in contributions.items():
    print(f"- {signal.capitalize()}: {value}%") 
```
